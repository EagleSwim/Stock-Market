suppuse we want to fill all the null in type2 so that every pokemon will have 2 type basen on the probabilites distribution of the 
type2

```python
type2_count = df['type2'].count()
type2_value_count = df['type2'].value_counts()
type_probabilities = type2_value_count.map(lambda amount: amount / type2_count)
type_probabilities
```

```python
# Randomly select a type based on probabilities
random_type = np.random.choice(type_probabilities.index, p=type_probabilities.values)
type_probabilities.index
```

```python
df['type2'] = df['type2'].apply(lambda x: np.random.choice(type_probabilities.index, p=type_probabilities.values) if pd.isnull(x) else x)
```
